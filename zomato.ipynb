```python
country_name=zomato_df.Country.value_counts().index
```

```python
country_count=zomato_df.Country.value_counts().values
```

```python
plt.pie(country_count[0:3],labels=country_name[0:3],autopct="%1.2f%%")
```

top 3 countries which uses zomato are India,United Kingdom,United States

```python
zomato_df.columns
```

```python
ratings=zomato_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"counts"})
```


  When Rating is between 4.5 to 4.9---> Excellent
When Rating are between 4.0 to 3.4--->very good
when Rating is between 3.5 to 3.9----> good
when Rating is between 3.0 to 3.4----> average
when Rating is between 2.5 to 2.9----> average
when Rating is between 2.0 to 2.4----> Poor




```python
ratings.columns
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='counts',data=ratings)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='counts',hue='Rating color',data=ratings)
```

```python
sns.countplot(x=ratings['Rating color'])
```

```python
zomato_df[zomato_df['Rating color']=="White"].groupby(["Country"]).size().reset_index().rename(columns={0:"count"}).sort_values(by='count',ascending=False)
```
